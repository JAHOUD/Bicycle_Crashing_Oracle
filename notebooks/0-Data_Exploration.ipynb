```python
#Import des librairies Python
import pandas as pd
import numpy as np
```

```python
#Chargement des fichiers 2019
df_carac_2019 = pd.read_csv('2019/caracteristiques-2019.csv', sep = ';', index_col=0)
df_lieux_2019 = pd.read_csv('2019/lieux-2019.csv', sep = ';')
df_usagers_2019 = pd.read_csv('2019/usagers-2019.csv', sep = ';')
df_vehicules_2019 = pd.read_csv('2019/vehicules-2019.csv', sep = ';', index_col=1)

#Chargement des fichiers 2020
df_carac_2020 = pd.read_csv('2020/caracteristiques-2020.csv', sep = ';', index_col=0)
df_lieux_2020 = pd.read_csv('2020/lieux-2020.csv', sep = ';')
df_usagers_2020 = pd.read_csv('2020/usagers-2020.csv', sep = ';')
df_vehicules_2020 = pd.read_csv('2020/vehicules-2020.csv', sep = ';', index_col=1)
```

## Fichier caracteristiques

```python
#Modifications des -1 sur les variables catégorielles
cols = ['lum','agg','int','atm','col']
df_carac_2020[cols] = df_carac_2020[cols].replace({-1:np.nan})

#Modifications sur latitude et longitude
df_carac_2020.lat = df_carac_2020.lat.str.replace(',','.')
df_carac_2020.long = df_carac_2020.long.str.replace(',','.')
df_carac_2020.lat = df_carac_2020.lat.astype('float')
df_carac_2020.long = df_carac_2020.long.astype('float')

#display(df_carac_2020.info())

display(df_carac_2020.head())

print("Pourcentage de NaN :\n")
print(round(df_carac_2020.isna().sum()/df_carac_2020.shape[0]*100,1))
```

## Fichier lieux

```python
#Modifications des -1 sur les variables catégorielles
cols = ['catr','circ','vosp','prof','plan','surf','infra','situ','nbv','vma']
df_lieux_2020[cols] = df_lieux_2020[cols].replace({-1:np.nan})

#Cas particulier de larrout, qui est un float mal renseigné
df_lieux_2020['larrout'] = df_lieux_2020['larrout'].replace({' -1':np.nan})
df_lieux_2020.larrout = df_lieux_2020.larrout.str.replace(',','.')
df_lieux_2020.larrout = df_lieux_2020.larrout.astype('float')

#Cas particulier de pr et pr1, avec des valeurs '(1)'
df_lieux_2020[['pr','pr1']] = df_lieux_2020[['pr','pr1']].replace({'(1)':np.nan})

#display(df_lieux_2020.info())

display(df_lieux_2020.head())

print("Pourcentage de NaN :\n")
print(round(df_lieux_2020.isna().sum()/df_lieux_2020.shape[0]*100,1))
```

## Fichier usagers

```python
df_vehicules_2020.catv.value_counts().head()
```

```python
#Modifications des -1 sur les variables catégorielles
cols = ['place','catu','grav','sexe','trajet','secu1','secu2','secu3','locp','actp','etatp']
df_usagers_2020[cols] = df_usagers_2020[cols].replace({-1:np.nan,' -1':np.nan})

#display(df_usagers_2020.info())

display(df_usagers_2020.head())

print("Pourcentage de NaN :\n")
print(round(df_usagers_2020.isna().sum()/df_usagers_2020.shape[0]*100,1))
```

## Fichier vehicules

```python
df_vehicules_2020.catv.value_counts().head()
```

```python
#Modifications des -1 sur les variables catégorielles
cols = ['senc','catv','obs','obsm','choc','manv','motor']
df_vehicules_2020[cols] = df_vehicules_2020[cols].replace({-1:np.nan,' -1':np.nan})

#Apparemment les Num_Acc sont reconnus en tant que int
df_vehicules_2020.Num_Acc = df_vehicules_2020.Num_Acc.astype('string')

#display(df_vehicules_2020.info())

display(df_vehicules_2020.head())

print("Pourcentage de NaN :\n")
print(round(df_vehicules_2020.isna().sum()/df_vehicules_2020.shape[0]*100,1))
```

## Essai avec une map

Pour installer GeoPandas : https://geopandas.org/en/stable/getting_started.html  
Commande à faire dans la console Anaconda :  
> `conda install -c conda-forge geopandas`

```python
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame

df = pd.read_csv("Long_Lats.csv", delimiter=',', skiprows=0, low_memory=False)

geometry = [Point(xy) for xy in zip(df['Longitude'], df['Latitude'])]
gdf = GeoDataFrame(df, geometry=geometry)   

#this is a simple map that goes with geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf.plot(ax=world.plot(figsize=(10, 6)), marker='o', color='red', markersize=15);
```
