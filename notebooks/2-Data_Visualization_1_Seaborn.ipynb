```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/donnes_19_20_fusionnes.csv', sep=';')
```

# Graphique représentant la gravité de l'accident en fonction de la présence ou non d'une voie cyclable en agglomération

```python
#On limite nos données aux accidents ayant lieu en agglomération et impliquant un cycliste
df_agg = df[(df['agg'] == 2) & (df['catv'] == 7.0)]
df_agg['grav'].replace([1,2,3,4], ['indemne', 'Tué', 'Blessé hospitalisé','Blessé léger'], inplace = True)

total = float(len(df_agg))
fig = plt.figure(figsize=(15,15))
axes = fig.subplots(2, 2)
ncount = len(df_agg)
# 1ere figure sur le subplot 1
ax = sns.countplot(x=df_agg['vosp'][df_agg['vosp'] ==0], hue = df_agg['grav'],hue_order=['indemne', 'Blessé léger', 'Blessé hospitalisé','Tué'], palette = ('green', 'blue', 'orange', 'red'),ax = axes[0,0]);
plt.title("Gravité de l'accident en fonction de la présence d'une piste cyclable ou pas");
plt.xticks(np.arange(1), ['sans objet']);
ax.set_xlabel('sans objet')
ax.set_title('Gravité de l accident sur une voie ne comportant pas de piste cyclable')
ax.set_ylim(0,ncount)
ax.grid(linestyle = '--')
# Annotate % on the figure
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format(height/total*100),
            ha="center")

# 2ere figure sur le subplot 2
ax = sns.countplot(x=df_agg['vosp'][df_agg['vosp'] ==1], hue = df_agg['grav'], hue_order=['indemne', 'Blessé léger', 'Blessé hospitalisé','Tué'], palette = ('green', 'blue', 'orange', 'red'),ax = axes[0,1]);
ax.set_xlabel('Piste cyclable')
ax.set_title('Gravité de l accident sur une voie avec piste cyclable')
ax.grid(linestyle = '--')
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format(height/total*100),
            ha="center")

# 3ere figure sur le subplot 3    
ax = sns.countplot(x=df_agg['vosp'][df_agg['vosp'] ==2], hue = df_agg['grav'], hue_order=['indemne', 'Blessé léger', 'Blessé hospitalisé','Tué'],palette = ('green', 'blue', 'orange', 'red'),ax = axes[1,0]);
ax.set_xlabel('Bande cyclable')
ax.set_title('Gravité de l accident sur une voie avec bande cyclable')
ax.grid(linestyle = '--')
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format(height/total*100),
            ha="center")
            
# 4ere figure sur le subplot 4
ax= sns.countplot(x=df_agg['vosp'][df_agg['vosp'] ==3], hue = df_agg['grav'],hue_order=['indemne', 'Blessé léger', 'Blessé hospitalisé','Tué'], palette = ('green', 'blue', 'orange', 'red'),ax = axes[1,1]);
ax.set_xlabel('Voie réservée')
ax.set_title('Gravité de l accident sur une voie avec voie réservée aux cyclistes')
ax.grid(linestyle = '--')
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format(height/total*100),
            ha="center")

#La très grande majorité des accidents ont lieu lorsqu'il n'y a pas de piste cyclable. Cependant, le ratio de décès est plus important lorsqu'il n'y a pas voie vélo.
```

```python
import matplotlib.ticker as ticker

#On limite nos données aux accidents ayant lieu en agglomération et impliquant un cycliste
df_agg = df[(df['agg'] == 2) & (df['catv'] == 7.0)]
df_agg['grav'].replace([1,2,3,4], ['indemne', 'Tué', 'Blessé hospitalisé','Blessé léger'], inplace = True)

total = len(df_agg[df_agg['vosp'] == 0.0])
total2 = len(df_agg[df_agg['vosp'] == 1.0])
total3 = len(df_agg[df_agg['vosp'] == 2.0])
total4 = len(df_agg[df_agg['vosp'] == 3.0])
fig = plt.figure(figsize=(15,15))
axes = fig.subplots(2, 2)
ncount = len(df_agg)
# 1ere figure sur le subplot 1
ax = sns.countplot(x=df_agg['vosp'][df_agg['vosp'] ==0], hue = df_agg['grav'],hue_order=['indemne', 'Blessé léger', 'Blessé hospitalisé','Tué'], palette = ('green', 'blue', 'orange', 'red'),ax = axes[0,0]);
plt.title("Gravité de l'accident en fonction de la présence d'une piste cyclable ou pas");
plt.xticks(np.arange(1), ['sans objet']);
ax.set_xlabel('sans objet')
ax.set_title('Gravité de l accident sur une voie ne comportant pas de piste cyclable')
ax.grid(linestyle = '--')
# Annotate % on the figure
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format(height/total*100),
            ha="center")

# 2ere figure sur le subplot 2
ax = sns.countplot(x=df_agg['vosp'][df_agg['vosp'] ==1], hue = df_agg['grav'], hue_order=['indemne', 'Blessé léger', 'Blessé hospitalisé','Tué'], palette = ('green', 'blue', 'orange', 'red'),ax = axes[0,1]);
ax.set_xlabel('Piste cyclable')
ax.set_title('Gravité de l accident sur une voie avec piste cyclable')
ax.grid(linestyle = '--')
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format(height/total2*100),
            ha="center")

# 3ere figure sur le subplot 3    
ax = sns.countplot(x=df_agg['vosp'][df_agg['vosp'] ==2], hue = df_agg['grav'], hue_order=['indemne', 'Blessé léger', 'Blessé hospitalisé','Tué'],palette = ('green', 'blue', 'orange', 'red'),ax = axes[1,0]);
ax.set_xlabel('Bande cyclable')
ax.set_title('Gravité de l accident sur une voie avec bande cyclable')
ax.grid(linestyle = '--')
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format(height/total3*100),
            ha="center")
            
# 4ere figure sur le subplot 4
ax= sns.countplot(x=df_agg['vosp'][df_agg['vosp'] ==3], hue = df_agg['grav'],hue_order=['indemne', 'Blessé léger', 'Blessé hospitalisé','Tué'], palette = ('green', 'blue', 'orange', 'red'),ax = axes[1,1]);
ax.set_xlabel('Voie réservée')
ax.set_title('Gravité de l accident sur une voie avec voie réservée aux cyclistes')
ax.grid(linestyle = '--')
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format(height/total4*100),
            ha="center")
```

- Le nombre d'accidents sans présence de piste cyclable est beaucoup plus importante (80k accidents contre environ 8500)
- Dans tous les cas, la part d'indemnes et de blessés légers est prépondérante (ce qui parait normal)
- Lorsqu'il n'y a pas de piste cyclable sur la voie, la part d'indemne baisse d'environ 10%, donc la gravité augmente.
- Cette diminution d'indemnes se retrouve par l'augmentation de la part de blessé légers (environ + 8% à 10%)
- Lorsqu'il n'y a pas de voie cyclable, la part des accidents avec blessé hosptitalisé est environ 3% plus importante.
- la part des accidents avec décès est stable peu importe qu'il y ait une voie cyclable ou non et se situe autour de 1%.

## Graphique représentant la gravité de l'accident en fonction de la catégorie de véhicule

```python
#On limite nos données aux accidents impliquant les vélos, les véhicules légers ainsi que les motos
df3  = df[(df['catv'] == 1.0) | (df['catv'] == 7.0)| (df['catv'] == 30.0) | (df['catv'] == 31.0) | (df['catv'] == 32.0) | (df['catv'] == 33.0) | (df['catv'] == 34.0)]

df3.catv.replace([1.0,7.0,2.0,30.0,31.0,32.0,33.0,34.0], 
    ['Vélos','Véhicules légers','2 roues','2 roues','2 roues','2 roues','2 roues','2 roues'], inplace=True)
df3['grav'].replace([1,2,3,4], ['indemne', 'Tué', 'Blessé hospitalisé','Blessé léger'], inplace = True)

total_ = len(df3[df3['catv'] == 'Véhicules légers'])
total2_ = len(df3[df3['catv'] == 'Vélos'])
total3_ = len(df3[df3['catv'] == '2 roues'])

fig = plt.figure(figsize=(15,15))
axes = fig.subplots(2, 2)

#1ere figure du subplot
axe = sns.countplot(x= df3['catv'][df3['catv'] == 'Véhicules légers'], hue = df3['grav'][df3['catv'] == 'Véhicules légers'], hue_order = ['indemne', 'Blessé léger', 'Blessé hospitalisé','Tué'], palette = ('green', 'blue', 'orange', 'red'), ax= axes[0,0]);
axe.set_title('Gravité de l accident si véhicule léger')
axe.grid(linestyle = '--')
for p in axe.patches:
    height = p.get_height()
    axe.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format(height/total_*100),
            ha="center")

#2eme figure du subplot
axe = sns.countplot(x= df3['catv'][df3['catv'] == 'Vélos'], hue = df3['grav'][df3['catv'] == 'Vélos'],hue_order = ['indemne', 'Blessé léger', 'Blessé hospitalisé','Tué'], palette = ('green', 'blue', 'orange', 'red'), ax= axes[0,1]);
axe.set_title('Gravité de l accident si vélo')
axe.grid(linestyle = '--')
for p in axe.patches:
    height = p.get_height()
    axe.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format(height/total2_*100),
            ha="center")

#3eme figure du subplot
axe = sns.countplot(x= df3['catv'][df3['catv'] == '2 roues'], hue = df3['grav'][df3['catv'] == '2 roues'], hue_order = ['indemne', 'Blessé léger', 'Blessé hospitalisé','Tué'], palette = ('green', 'blue', 'orange', 'red'), ax= axes[1,0]);
axe.set_title('Gravité de l accident si 2 roues')
axe.grid(linestyle = '--')
for p in axe.patches:
    height = p.get_height()
    axe.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format(height/total3_*100),
            ha="center")
```

- La part d'indemnes pour les accidents impliquant les véhicules légers est beaucoup plus importante
- Ceci peut être interprété car les accidents bénins impliquant des vélos et 2 roues sont beaucoup moins recensés
- Cependant, la part des accidents avec des blessés légers et hospitalisés est beaucoup plus importantes pour les 2 roues et les vélos
- La part des tués est légèrement plus importante pour les 2 roues et les vélos (seulement 1% de plus)
